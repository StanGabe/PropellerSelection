```python
import numpy as np
#Compute KQ
def calculate_Y1(C1, O1, a1, h1, D1, X1, E1, G1, L1):
    m = len(O1)
    n = 4
    # Calculate the inner sum
    inner_sum = np.zeros(m)
    for i in range(m):
        inner_sum[i] = a1[i]
        for k in range(n):
            inner_sum[i] += h1[k][i] * (D1[k] * X1[k] + E1[k])
    
    # Calculate the outer sum
    outer_sum = C1 + np.sum(O1 * np.tanh(inner_sum))
    
    # Calculate Y
    Y = (np.tanh(outer_sum) - G1) / L1
    
    return Y

#Compute KT
def calculate_Y2(C2, O2, a2, h2, D2, X2, E2, G2, L2):
    m = len(O2)
    n = 4
    # Calculate the inner sum
    inner_sum = np.zeros(m)
    for i in range(m):
        inner_sum[i] = a2[i]
        for k in range(n):
            inner_sum[i] += h2[k][i] * (D2[k] * X2[k] + E2[k])
    
    # Calculate the outer sum
    outer_sum = C2 + np.sum(O2 * np.tanh(inner_sum))
    
    # Calculate Y
    Y = (np.tanh(outer_sum) - G2) / L2
    
    return Y
```

```python
import pyswarms as ps
n_bounds = (3000,3500)
def objective_1(X):
    n_particles = X.shape[0]  # number of particles
    output = np.zeros(n_particles)  # Initialize array for storing outputs

    for i in range(n_particles):
        J, n, D, pod = X[i]
        V = J * n * D/60
        patm=101300 
        rho=1025 
        g=9.81 
        h=1
        po=patm+rho*g*h
        pv =3200 
        sigma= (po-pv)/(0.5*rho*V**2)
        X1[1] = 0.8
        X1[2] = pod
        X1[3] = sigma
        X1[0] = J 
        KT_1 = calculate_Y2(C2, O2, a2, h2, D2, X1, E2, G2, L2)
        KQ_1 = calculate_Y1(C1, O1, a1, h1, D1, X1, E1, G1, L1)
        T = (KT_1 * rho * n**2 * D**4)/3600
        P = (KQ_1 * rho * n**3 * D**5*2*np.pi)/216000
        R = spline(V) 
        if R > T+0.1*R or R < T -0.1*R:
            output[i] = 1e12  # Return a large value for excessive thrust
        elif P > Pmax:
            output[i] = 1e12  # Return a large negative value for excessive power
        else:
            output[i] = -V  # Negative sign to maximize speed

    return output

lb = [J_bounds[0], n_bounds[0], D_bounds[0], pod_bounds[0]]
ub = [J_bounds[1], n_bounds[1], D_bounds[1], pod_bounds[1]]

options = {'c1': 0.5, 'c2': 0.3, 'w':0.9}

optimizer = ps.single.GlobalBestPSO(n_particles=20, dimensions=4, options=options, bounds=(lb, ub))

cost, pos = optimizer.optimize(objective_1, iters=300)
```
