```python
import numpy as np
#Compute KQ
def calculate_Y1(C1, O1, a1, h1, D1, X1, E1, G1, L1):
    m = len(O1)
    n = 4
    # Calculate the inner sum
    inner_sum = np.zeros(m)
    for i in range(m):
        inner_sum[i] = a1[i]
        for k in range(n):
            inner_sum[i] += h1[k][i] * (D1[k] * X1[k] + E1[k])
    
    # Calculate the outer sum
    outer_sum = C1 + np.sum(O1 * np.tanh(inner_sum))
    
    # Calculate Y
    Y = (np.tanh(outer_sum) - G1) / L1
    
    return Y

#Compute KT
def calculate_Y2(C2, O2, a2, h2, D2, X2, E2, G2, L2):
    m = len(O2)
    n = 4
    # Calculate the inner sum
    inner_sum = np.zeros(m)
    for i in range(m):
        inner_sum[i] = a2[i]
        for k in range(n):
            inner_sum[i] += h2[k][i] * (D2[k] * X2[k] + E2[k])
    
    # Calculate the outer sum
    outer_sum = C2 + np.sum(O2 * np.tanh(inner_sum))
    
    # Calculate Y
    Y = (np.tanh(outer_sum) - G2) / L2
    
    return Y
```

```python
import cupy as cp

#Compute KQ
def calculate_Y1_cupy(C1, O1, a1, h1, D1, X1, E1, G1, L1):
    X11 = cp.asarray(X1)
    C11 = cp.asarray(C1)
    O11 = cp.asarray(O1)
    a11 = cp.asarray(a1)
    h11 = cp.asarray(h1)
    D11 = cp.asarray(D1)
    E11 = cp.asarray(E1)
    G11 = cp.asarray(G1)
    L11 = cp.asarray(L1)
    m = len(O1)
    n = 4
    # Calculate the inner sum
    inner_sum = cp.zeros(m)
    for i in range(m):
        inner_sum[i] = a11[i]
        for k in range(n):
            inner_sum[i] += h11[k][i] * (D11[k] * X11[k] + E11[k])
    
    # Calculate the outer sum
    outer_sum = C11 + cp.sum(O11 * cp.tanh(inner_sum))
    
    # Calculate Y
    Y = (cp.tanh(outer_sum) - G11) / L11
    
    return Y

#Compute KT
def calculate_Y2_cupy(C2, O2, a2, h2, D2, X2, E2, G2, L2):
    X22 = cp.asarray(X2)
    C22 = cp.asarray(C2)
    O22 = cp.asarray(O2)
    a22 = cp.asarray(a2)
    h22 = cp.asarray(h2)
    D22 = cp.asarray(D2)
    E22 = cp.asarray(E2)
    G22 = cp.asarray(G2)
    L22 = cp.asarray(L2)
    m = len(O2)
    n = 4
    # Calculate the inner sum
    inner_sum = cp.zeros(m)
    for i in range(m):
        inner_sum[i] = a22[i]
        for k in range(n):
            inner_sum[i] += h22[k][i] * (D22[k] * X22[k] + E22[k])
    
    # Calculate the outer sum
    outer_sum = C22 + cp.sum(O22 * cp.tanh(inner_sum))
    
    # Calculate Y
    Y = (cp.tanh(outer_sum) - G22) / L22
    
    return Y

```

```python
import numpy as np
import matplotlib.pyplot as plt

def differential_evolution(func, bounds, popsize=40, its=100):
    num_params = len(bounds)
    pop = np.random.rand(popsize, num_params)
    min_b, max_b = np.asarray(bounds).T
    diff = np.fabs(min_b - max_b)
    population = min_b + pop * diff
    fitness = np.asarray([func(ind) for ind in population])
    best_idx = np.argmin(fitness)
    best = population[best_idx]
    pop_history = []  # Create a list to store the population history
    pop_1 = []
    for i in range(its):
        for j in range(popsize):
            idxs = [idx for idx in range(popsize) if idx != j]
            a, b, c = population[np.random.choice(idxs, 3, replace = False)]
            mutant = np.clip(a + 0.5 * (b - c), min_b, max_b)
            cross_points = np.random.rand(num_params) < 0.9
            if not np.any(cross_points):
                cross_points[np.random.randint(0, num_params)] = True
            trial = np.where(cross_points, mutant, population[j])
            trial_fitness = func(trial)
            if trial_fitness < fitness[j]:
                fitness[j] = trial_fitness
                population[j] = trial
                if trial_fitness < fitness[best_idx]:
                    best_idx = j
                    best = trial
        if i % 4 == 0:
            pop_1.append(np.array(population))
            pop_history.append(np.prod(population[:, :3], axis=1)/60)  # Store the product of each individual in the population
            
    return best, fitness[best_idx], pop_history,pop_1  # Also return the population history

bounds = [J_bounds, n_bounds, D_bounds, pod_bounds]
result, fitness, pop_history,pop1 = differential_evolution(objective, bounds)  # Get the population history from the function
```
